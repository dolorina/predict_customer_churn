```python
obj = df.isnull().sum()
columns, index = [], []
for key, val in obj.iteritems():
    index.append(key)
    columns.append(val)

isnull_sum= pd.DataFrame(index=index)
for cols in columns:
    isnull_sum[cols] = pd.Series(dtype=int)
    isnull_sum[cols] = cols

dfi.export(isnull_sum, 'isnull_sum.png')
```
